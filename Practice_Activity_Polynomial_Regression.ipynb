```python
import sklearn
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline

lr = LinearRegression()

#create X and y variables
y = df['bill_depth_mm']
X = df[['bill_length_mm']]
#fit the model
lr_fit = lr.fit(X, y)
#predict the values in the test data
y_pred_lr = lr_fit.predict(X)

y_pred_lr
```

```python
#polynomial model

poly_2 = make_pipeline(PolynomialFeatures((1,2)), LinearRegression())

#fit the model
m2 = poly_2.fit(X,y)
#predict the values in the test data
y_pred_quad = m2.predict(X)

y_pred_quad


```

```python
#polynomial 3 model

poly_3 = make_pipeline(PolynomialFeatures((1,3)), LinearRegression())

#fit the model
m3 = poly_3.fit(X,y)
#predict the values in the test data
y_pred_cube = m3.predict(X)

y_pred_cube
```

```python
#Degree 10 polynomial model

poly_10 = make_pipeline(PolynomialFeatures((1,10)), LinearRegression())

#fit the model
m4 = poly_10.fit(X,y)
#predict the values in the test data
y_pred_ten = m4.predict(X)

y_pred_ten
```
